```python
vocab_size = 50257
context_length = 1024
num_layers = 48
d_model = 1600
num_heads = 25
d_ff = 6400

d_k = d_model // num_heads * num_heads

embedding_params = vocab_size * d_model * 2
ln_final_params = d_model
layers_params = (d_model + d_model * d_k * 4 + d_model + d_model * d_ff * 3) * num_layers
sum_params = embedding_params + ln_final_params + layers_params
print(f'Total params: {sum_params}, memory required: {sum_params * 4 / 1e9} GB')
```

```python
def print_float_op(num_layers, d_model, num_heads):
    d_k = d_model // num_heads * num_heads
    op_mha = 2 * d_model * d_k * context_length * 4 + 2 * context_length * context_length * d_k * 2
    op_swiglu = 2 * d_model * d_ff * context_length * 3
    op_per_layer = op_mha + op_swiglu
    op_layers = op_per_layer * num_layers
    op_lm_head = 2 * d_model * vocab_size * context_length
    op_sum = op_layers + op_lm_head
    print(f'op_mha {op_mha * num_layers / 1e9} G float operations')
    print(f'op_swiglu {op_swiglu * num_layers / 1e9} G float operations')
    print(f'op_lm_head {op_lm_head / 1e9} G float operations')
    print(f'op_sum {op_sum / 1e9} G float operations')
print_float_op(num_layers, d_model, num_heads)
```
